```python
# ì•„ë˜ì— ì½”ë“œë¥¼ ì‘ì„±í•˜ì„¸ìš”.

class Pokemon:
    level = 5
    exp = 0
    hp = level * 20
    pika_list = []
    def __init__(self,name):
        self.name = name
        self.level = Pokemon.level
        self.hp = Pokemon.hp
        self.exp = Pokemon.exp
        Pokemon.pika_list.append(name)
        
    def GetStatus(self):
        name = self.name
        level = self.level
        hp = self.hp
        exp = self.exp
        print(f"ì´ë¦„ : {name}, ë ˆë²¨ : {level}, HP : {hp}, EXP : {exp}")
        
    def bark(self):
        #d1 = Pokemon('d1')# í´ë˜ìŠ¤ ì•ˆì—ì„œ ë™ì¼ í´ë˜ìŠ¤ ìƒì„±ì´ ë˜ë„¤...
        return print(f'í”¼ì¹´í”¼ì¹´!! ë‚˜ëŠ” {self.name}!!') #,Pokemon.pika_list,type(d1))
    
                #self , enemy ë‘˜ë‹¤ Pokemonì´ë¼ëŠ” ì „ì œ
    def body_attack(self,enemy):
        # ìƒëŒ€ë°©ì˜ hpë¥¼ ê¹ì•„ì•¼í•œë‹¤.
        # ìƒëŒ€ë°© ê°ì²´ê°€ ì´ ë©”ì„œë“œ ì•ˆì—ì„œ ì‚¬ìš©ë˜ì–´ì•¼í•¨...
        # ì–´ë–»ê²Œë“  ìƒëŒ€ë°© ê°ì²´ë¥¼ ëŒê³  ë“¤ì–´ì™€ì•¼í•œë‹¤.
        # ë”ë¯¸ë¥¼ ë§Œë“œë‚˜? dummy = Pokemon('dummy') ì´ë ‡ê²Œ?
        # ê·¸ëƒ¥ ë³€ìˆ˜ ë‘ê°œ ë°›ëŠ” ê±°ì˜€ë‹¤....(self, enemy) ì´ë ‡ê²Œ...
        power = self.level * 5
        enemy.hp -= power 
        print(f"{self.name}ì˜ ëª¸í†µë°•ì¹˜ê¸°! {power}ì˜ í”¼í•´ë¥¼ ì…í˜”ë‹¤!")
        
        if enemy.hp <= 0:
            self.exp += enemy.level * 15
            self.IsLevelUp()
            print(f"ì•„ì´ê³  {enemy.name} ì£½ë„¤!!")
            del enemy
    
    def thousond_volt(self,enemy):
        power = self.level * 7
        enemy.hp -= power
        print(f"{self.name}ì˜ 10ë§Œë³¼íŠ¸!! {power}ì˜ í”¼í•´ë¥¼ ì…í˜”ë‹¤!")
        
        if enemy.hp <= 0:
            self.exp += enemy.level * 15
            self.IsLevelUp()
            print(f"ì•„ì´ê³  {enemy.name} ì£½ë„¤!!")
            del enemy

    # ì´ë ‡ê²Œ ì ì ‘ê·¼ì„ í•˜ë©´ ì›ë³¸ì„ ë°”ê¿€ìˆ˜ ìˆë‹¤?
    # ì§„ì§œë¡œ ë°”ê¿€ìˆ˜ ìˆë‹¤. ë­”ê°€ í•˜ë©´ì„œë„ ì–´ì•ˆì´ ë²™ë²™í•˜ë‹¤...
    def IsLevelUp(self):
        if self.exp >= self.level * 100:
            self.exp -= self.level * 100
            self.level += 1
            self.hp = self.level * 20
        else:
            pass
        
```

```python
# (ì„ íƒì  ë„ì „ê³¼ì œ) ì „íˆ¬ ì½”ë“œ ì˜ˆì‹œ

'''while ë‘ í¬ì¼“ëª¬ ì¤‘ í•˜ë‚˜ê°€ ì“°ëŸ¬ì§€ê¸° ì „ê¹Œì§€:
    1. randomì„ ì´ìš©í•˜ì—¬ ë¬´ì‘ìœ„ë¡œ ê³µê²©ê¶Œ ì„ ì • (A = í™€/p1 ì§/p2)
    1.1 choiceë¡œ ë‘˜ì¤‘ í•˜ë‚˜ ê³ ë¦„ , ê°ì²´ì— í• ë‹¹ (A)
    2. ê³µê²©ê¶Œ ê°€ì§„ í¬ì¼“ëª¬ì´ ìƒëŒ€ë°© ê³µê²©! (B)
    2.1 B%3--> 0: ìš¸ê¸° 1 : ëª¸í†µë°•ì¹˜ê¸° 2 : 10ë§Œë³¼íŠ¸)
    3. ë‘˜ ì¤‘ í•˜ë‚˜ê°€ ì“°ëŸ¬ì§ˆ ë•Œê¹Œì§€ 1~2ë²ˆ ê³„ì† ë°˜ë³µ'''

import random

p1 = Pokemon('ê¹€í”¼ì¹´')
p2 = Pokemon('ë°•í”¼ì¹´')
Pokemon_list = ['p1','p2']

print("ì‹œí•© ì‹œì‘!")
print("")

while (p1.hp>0 and p2.hp>0):
    A = random.sample(Pokemon_list,1) #sampleë¡œ í•˜ë‹ˆê¹Œ ê¹Œë‹¤ë¡­ë„¤... ì–˜ê°€ listë¡œ ë±‰ì–´ì£¼ë‹ˆê¹Œ, ì´í•˜ì˜ ifë¬¸ ì“°ë ¤ë©´ ë¦¬ìŠ¤íŠ¸ ì–‘ì‹ê¹Œì§€ ë§ì¶°ì•¼í•œë‹¤.. 
    #A = random.randint(1,100)
    B = random.randint(1,100)
    print(A,B)
    if A == ['p1'] : # p1 ê³µê²©(ì§ìˆ˜)
        if B%3 == 0:
            p1.bark()
        elif B%3 == 1:
            p1.body_attack(p2)
        else:
            p1.thousond_volt(p2)
    else: #p2 ê³µê²© (í™€ìˆ˜)
        if B%3 == 0:
            p2.bark()
        elif B%3 == 1:
            p2.body_attack(p1)
            
        else:
            p2.thousond_volt(p1)        
print("")
print("ì‹œí•© ë!")
    

```

```python
# êµìˆ˜ë‹˜ ì½”ë“œ

import random

p1 = Pokemon('ìš©ì¬ì¹´ì¸„')
p2 = Pokemon('ë¯¸ì •ì¹´ì¸„')

# í•¨ìˆ˜í™”ë¡œ ì¬ì‚¬ìš©ì„±ë„ ë†’ì„
def battle(p1, p2):
    print('ê²½ê¸°... ì‹œì‘! ğŸ”¥')
    print()
    
    skills = ['bark', 'body_attack', 'thousand_volt']
    while p1.hp > 0 and p2.hp > 0:
        
        # ì•„ê¹Œ ë‚˜ëŠ” choiceê°€ ë¦¬ìŠ¤íŠ¸ë¡œ ë±‰ì–´ì¤€ë‹¤ê³  í•´ì„œ ì–´ë µë‹¤ê³  í–ˆëŠ”ë°,
        # ë¯¸ë¦¬ ë¦¬ìŠ¤íŠ¸ë¡œ ì²˜ë¦¬í•´ì„œ 'p1' 'p2'ë¥¼ ë±‰ì–´ë‚´ë„ë¡ í–ˆë„¤
        attacker = random.choice([p1, p2])
        
        # *** ë³€ìˆ˜ì— ì¸ìŠ¤í„´ìŠ¤ë¥¼ í†µì§¸ë¡œ ë„£ì„ìˆ˜ ìˆë‹¤ê³ ???
        # ì•„ë‹ˆ ìƒê°í•´ë³´ë©´ ë§¨ë‚  í•˜ë˜ ì¼ì´ì–ì•„.. a = Play_list ì´ëŸ° ì‹ìœ¼ë¡œ ë³€ìˆ˜ í• ë‹¹í•˜ë“¯ í•˜ëŠ” ì¼ì´êµ°
        # ê·¸ë˜ë„ ì´ë ‡ê²Œê¹Œì§€ ìœ ì—°í• ì¤„ì´ì•¼... ê·¸ë˜ì„œ íŒŒì´ì¬ì´ ê°ì²´ ì¤‘ì‹¬ ì–¸ì–´ë¼ëŠ” ê²ƒì¸ê°€?
        defender = p1 if attacker == p2 else p2
        
        #skillsì—ì„œ ëœë¤í•˜ê²Œ ë¬¸ìì—´ì„ ë½‘ì•„ì˜´
        skill = random.choice(skills)
        
        # ë¬¸ìì—´ê³¼ ë§ëŠ” ì¸ìŠ¤í„´ìŠ¤ ë©”ì†Œë“œë¥¼ ì‹¤í–‰ì‹œí‚´...
        if skill == 'bark':
            attacker.bark()
        elif skill == 'body_attack':
            attacker.body_attack(defender)
        elif skill == 'thousand_volt':
            attacker.thousand_volt(defender)
    
    print()
    print('ê²½ê¸° ì¢…ë£Œ!')
```
